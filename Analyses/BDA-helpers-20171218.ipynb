## Fitting theta

```python
p = '/Users/benpeloquin/Desktop/Projects/rsa_uid/Data/theta-fitting.csv'
df = pd.read_csv(p)
```

```python
def createData(df):
    bda_data = []
    for (supr, total, nMarked) in list(zip(df['supr'], df['total'], df['numMarked'])):
        points = ['X a' for _ in range(nMarked)] + ['a' for _ in range(total-nMarked)]
        semProbs = [np.exp(supr) for _ in range(total)]
        for point, semProb in zip(points, semProbs):
            bda_data.append({
                'semProb': semProb,
                'marked': point
            })
    return bda_data
    
data = createData(df)
```
